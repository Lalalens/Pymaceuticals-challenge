```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import pearsonr, linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_mouse_study = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")

# Display the data table for preview
merged_mouse_study.head()
```

```python
# Checking the number of mice.
total_mice = len(mouse_metadata["Mouse ID"])
total_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
multi_mice = merged_mouse_study[merged_mouse_study.duplicated(["Mouse ID", "Timepoint"])]
multi_mice_ids = multi_mice["Mouse ID"].unique()
multi_mice_ids
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
all_multi_mice = merged_mouse_study[merged_mouse_study["Mouse ID"].isin(multi_mice_ids)]
all_multi_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_mouse_study = merged_mouse_study.drop_duplicates("Mouse ID")
cleaned_mouse_study.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# cleaned_mouse_study.describe()
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

mean_tum_vol = merged_mouse_study.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
med_tum_vol = merged_mouse_study.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
tum_vol_var = merged_mouse_study.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
tum_vol_std = merged_mouse_study.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
tum_vol_sem = merged_mouse_study.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

mice_summary_study = pd.DataFrame({
    "Mean Tumor Volume": mean_tum_vol, 
    "Median Tumor Volume": med_tum_vol,
    "Tumor Volume Variance": tum_vol_var,
    "Tumor Volume Std. Dev.": tum_vol_std,
    "Tumor Volume Std. Err.": tum_vol_sem
})

mice_summary_study
 
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
mice_summary_study = merged_mouse_study.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(
    mean=("mean"),
    median=("median"),
    var=("var"),
    std=("std"),
    sem=("sem"),
)

mice_summary_study
```

## Quartiles, Outlires and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

drug_treat = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
filtered_drug_treat = merged_mouse_study[merged_mouse_study["Drug Regimen"].isin(drug_treat)]

# Start by getting the last (greatest) timepoint for each mouse
great_timepoints = filtered_drug_treat.groupby("Mouse ID")["Timepoint"].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_tum_vol = pd.merge(great_timepoints, 
                             filtered_drug_treat, on=["Mouse ID", "Timepoint"], how="left")

merged_tum_vol.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_treatment = merged_tum_vol[merged_tum_vol
                                        ["Drug Regimen"] == treatment]["Tumor Volume (mm3)"]
    
    # add subset
    tumor_vol_data.append(tumor_treatment)
    
    #Obtain IQR
    quartile = tumor_treatment.quantile([0.25, 0.5, 0.75])
    lowerq = quartile[0.25]
    upperq = quartile[0.75]
    iqr = upperq - lowerq
    
     
    # Determine outliers using upper and lower bounds
    lower_bond = lowerq - (1.5 * iqr)
    upper_bond = upperq - (1.5 * iqr)
    
    outliers = tumor_treatment[(tumor_treatment < lower_bond) | (tumor_treatment > upper_bond)]
    
    print(f"{treatment}'s potential outliers: {outliers.to_string(index=False)}{tumor_treatment}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
treatment_labels = treatments

fig1, ax1 = plt.subplots()
ax1.boxplot(tumor_vol_data, labels=treatment_labels)

ax1.set_title("Final Tumor Volume (mm3)")
ax1.set_xlabel("Treatment Group")
ax1.set_ylabel("Tumore Volume (mm3)")

plt.show()
```

## Correlation and Regression

```python
capomulin_data = merged_tum_vol.loc[merged_tum_vol["Drug Regimen"] == "Capomulin"]
grouped_data = capomulin_data.groupby("Mouse ID").agg({"Tumor Volume (mm3)": 
                                                       "mean", "Weight (g)": "mean"})

# Calculate the correlation coefficient
correlation_coefficient, _ = pearsonr(grouped_data['Weight (g)'], grouped_data['Tumor Volume (mm3)'])
print(f"The correlation between mouse weight and average tumor volume is: {correlation_coefficient:.2f}")

(slope, intercept, rvalue, pvalue, stderr) = linregress(grouped_data["Weight (g)"], 
                                                       grouped_data["Tumor Volume (mm3)"])
line_eq = f"y = {slope:.2f}x + {intercept:.2f}"

grouped_data["Predicted Tumor Volume"] = slope * grouped_data["Weight (g)"] + intercept

plt.scatter(grouped_data["Weight (g)"], grouped_data["Tumor Volume (mm3)"])
plt.plot(grouped_data["Weight (g)"], grouped_data["Predicted Tumor Volume"], color="purple")

plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.annotate(line_eq, (20,36), color="purple")

plt.show()
```
